## **Importing Packages**

```python
import pyshorteners as pyshort
```

## **URL Shortening Function**

```python
def shortening(url):
  short_url=pyshort.Shortener()
  return short_url.tinyurl.short(url)
```

## **Input**

```python
url=input("Enter the link you need to shorten").strip()
print(shortening(url))
```
